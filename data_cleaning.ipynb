```python
import pandas as pd
```

```python
dataset = "ph_raw_data.csv"
insert_columns = ["embed", "thubnail_1", "thumbnail_2", "title", "tags", "categories", "pornstars", "duration", "views", "likes", "dislikes", "thumbnail_3", "thumbnail_4"]
```

```python
df = pd.read_csv(dataset, delimiter = "|", on_bad_lines = "skip", header = None)
df.columns = insert_columns
len(df)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Drop any row which contains NaN
df = df.dropna(subset = ["tags", "categories", "pornstars"])
```

```python
# Define a function to split values with ";" in columns and convert them into a list
def split_semicolon_values_and_convert(df, column_name):
    df[column_name] = df[column_name].str.split(";").apply(lambda x: [str(item) for item in x if not pd.isna(item)])
```

```python
# Apply function in "tags", "categories" and "pornstars"
split_semicolon_values_and_convert(df, "tags")
split_semicolon_values_and_convert(df, "categories")
split_semicolon_values_and_convert(df, "pornstars")
```

```python
# Save cleaned dataset in a new .csv file
df.to_csv("ph_cleaned_data_oct_31.csv", index = False)
```
