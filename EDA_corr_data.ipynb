```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("ph_cleaned_data_oct_31.csv")
df.head()
```

```python
# Duration analysis
```

```python
df["duration"]
```

```python
# Assumption: time in seconds --> transform to minutes
```

```python
df["duration"] = df["duration"].apply(lambda x: (x / 60)).round(2)
```

```python
# Mean duration

mean_duration = round(df["duration"].mean(), 2)
print(f"Mean duration is {mean_duration} minutes")
```

```python
# Median duration

median_duration = round(df["duration"].median(), 2)
print(f"Median duration is {median_duration} minutes")
```

```python
# Examine the 20 most viewed video titles and their duration

top_20_most_viewed_videos_duration = df.sort_values(by = "views", ascending = False).head(20)
top_20_most_viewed_videos_duration[["title", "duration"]]
```

```python
# Correlation between variables
```

```python
df[["duration", "views", "likes", "dislikes"]].corr()
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap = True) # Set custom colors; cmap = cmap must be set in sns.heatmap() as an argument

heatmap = sns.heatmap(df[["duration", "views", "likes", "dislikes"]]
            .corr(), annot = True, vmax = 1, center = 0, cmap = cmap, square = True, linewidths = .5, cbar_kws = {"shrink": .5})
heatmap.set_title("Linear correlation between variables")
```

```python
pairplot = sns.pairplot(df[["duration", "views", "likes", "dislikes"]], diag_kind = "kde")
```

```python
boxplot = df.boxplot(column = ["duration", "views", "likes", "dislikes"])
boxplot.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))
```

```python
# Like vs dislike analysis
```

```python
# Identify title with most likes

most_liked_row = df.loc[df["likes"].idxmax()]
print("Title with most likes:", most_liked_row["title"])
print("Number of likes:", most_liked_row["likes"])
```

```python
# Identify title with least likes

least_liked_row = df.loc[df["likes"].idxmin()]
print("Title with least likes:", least_liked_row["title"])
print("Number of likes:", least_liked_row["likes"])
```

```python
# Identify title with most dislikes
most_disliked_row = df.loc[df["dislikes"].idxmax()]
print("Title with most dislikes:", most_disliked_row["title"])
print("Number of dislikes:", most_liked_row["dislikes"])
```

```python
# Identify title with least dislikes

least_disliked_row = df.loc[df["dislikes"].idxmin()]
print("Title with least dislikes:", least_disliked_row["title"])
print("Number of dislikes:", least_disliked_row["dislikes"])
```

```python
# Define total likes and dislikes
total_likes = df["likes"].sum()
total_dislikes = df["dislikes"].sum()

# Parameter for chart
label = ["Likes", "Dislikes"]
size = [total_likes, total_dislikes]

# Create pie chart
fig, ax = plt.subplots()

ax.pie(size, labels = label, autopct = lambda p: "{:.0f}K"
       .format(p * sum(size) / 1000), startangle = 90, colors = ["springgreen", "tomato"])
plt.title("Total like-dislike-ratio across all titles in [%]")
plt.axis("equal") # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```

```python
# Define mean likes and dislikes (is equal to .sum() like-dislike-ration)
mean_likes = df["likes"].mean()
mean_dislikes = df["dislikes"].mean()

# Parameter for chart
mean_label = ["Mean Likes", "Mean Dislikes"]
mean_size = [mean_likes, mean_dislikes]

# Create pie chart
plt.pie(mean_size, labels = mean_label, autopct = "%1.1f%%", startangle = 90, colors = ["springgreen", "tomato"])
plt.title("Mean like-dislike-ratio across all titles in [%]")
plt.axis("equal") # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```

```python
# Define median likes and dislikes
median_likes = df["likes"].median()
median_dislikes = df["dislikes"].median()

# Parameter for chart
median_label = ["Median Likes", "Median Dislikes"]
median_size = [median_likes, median_dislikes]

# Create pie chart
plt.pie(median_size, labels = median_label, autopct = "%1.1f%%", startangle = 90, colors = ["springgreen", "tomato"])
plt.title("Median like-dislike-ratio across all titles in [%]")
plt.axis("equal") # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```
